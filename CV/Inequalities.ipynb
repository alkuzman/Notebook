# Hlder's Inequality



```python
def Lp_norm(x, p=2):
    return np.sum(np.abs(x) ** p) ** (1/p)

interval_size = 10000000
p = 2
q = 2
fx = np.random.randint(-100, 100, interval_size)
gx = np.random.randint(-100, 100, interval_size)

fg_l1 = Lp_norm(fx * gx, 1)
f_lp = Lp_norm(fx, p)
g_lq = Lp_norm(gx, q)

diff = f_lp * g_lq - fg_l1
 
print(diff/fg_l1)
```
