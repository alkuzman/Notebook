```python
import numpy as np
import sympy as sym
from matplotlib import pyplot as plt
# from sympy.plotting import plot3d

from ipywidgets import interact
from ipywidgets import FloatSlider

x, y = sym.symbols('x y')
%matplotlib widget
# %matplotlib inline
```

```python
def plot_homogeneus_line(theta, t, w_length):
    w = w_length * (sym.cos(theta) * x + sym.sin(theta) * y)
    
    sym.plot_implicit(sym.Eq(w, t), (x, -3, 3), (y, -3, 3))

    V = np.array([np.cos(theta), np.sin(theta)]) * w_length
    plt.quiver(0, 0, V[0], V[1], color='black',angles='xy', units='xy', scale_units='xy', scale=1, width=0.03) 
    
interact(
    plot_homogeneus_line,
    theta=FloatSlider(min=-np.pi, max=np.pi, step=0.01, value=np.pi/4, continuous_update=False),
    t=FloatSlider(min=-10, max=10, step=0.01, value=1, continuous_update=False),
    w_length=FloatSlider(min=0.01, max=5, step=0.01, value=1, continuous_update=False), continuous_update=False);  
```
