# Thermo

*Thermo* is small pet project that I started so I can gather **Temperature** and **Boiler State** data from my thermostat (which is not WiFi enabled). My thermostat is `Vaillant VRT330`.






This will give me information on what is the *achived temperature* and allow me to control the separate radiators on which I installed `Tado` devices according to the main temperature. Hopefully, I will be able to provide heating on demand in each room
