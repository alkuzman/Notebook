* Every deviation of this frame will be treated as foreground

### Background Subtraction

````python
def background_subtraction(in_stream, out_stream, threshold=0.2):
    thresh = threshold * 3 * 255
    background = None
    i = 0
    with in_stream as stream, out_stream as ostream:
        for ret, frame in stream:
            # Our operations on the frame come here
            im = frame.astype(np.float32)
            if background is None:
                background = im

            diff = (np.linalg.norm((background - im), axis=2) > thresh).astype(np.float32)

            mask, masked = cvu.create_mask(im, diff)

            # Lets the camera ligthning stabilize.
            if i < 20:
                background = im
            i += 1

            ostream((background, frame, mask, masked), (2, 2))

    # When everything is done, release the capture
````

````python
background_subtraction(cvu.cam_stream(), cvu.window_stream(), threshold=0.07)
````

## Sequential Frame Differencing

![](./sequential-frame-differencing.png)

````python
def frame_differencing(in_stream, out_stream, threshold=0.2):
    thresh = threshold * 3 * 255
    prev_frame = None
    with in_stream as stream, out_stream as ostream:
        for ret, frame in stream:

            # Our operations on the frame come here
            frame = frame.astype(np.float32)
            
            if prev_frame is None:
                prev_frame = frame

            diff = (np.linalg.norm((prev_frame - frame), axis=2) > thresh).astype(np.float32)

            prev_frame = frame
            
            mask, masked = cvu.create_mask(frame, diff)

            # Display the resulting frame
            ostream((prev_frame, frame, mask, masked), (2, 2))
````

````python
frame_differencing(cvu.cam_stream(), cvu.window_stream(), threshold=0.07)
````

### Delay improvement

![](./non-sequential-frame-differencing.png)

$$
D(K) = || I(t) - I(t-K) ||
$$

````python
def frame_differencing_with_delay(in_stream, out_stream, threshold=0.2, delay=10):
    thresh = threshold * 3 * 255
    index = -1

    prev_frames = [None] * (delay + 1)
    prev_indices = [None] * (delay + 1)
    
    with in_stream as stream, out_stream as ostream:
        for ret, frame in stream:
            index += 1

            # Our operations on the frame come here
            frame = frame.astype(np.float32)
            i = index % (delay + 1)
            prev_frames[i] = frame
            prev_indices[i] = index

            prev_index = max(0, index - delay)
            prev_i = prev_index % (delay + 1)
            ind = prev_indices[prev_i]
            prev_frame = prev_frames[prev_i]

            if prev_index != ind:
                diff = frame
            else:
                diff = (np.linalg.norm((prev_frame - frame), axis=2) > thresh).astype(np.float32)
                
            mask, masked = cvu.create_mask(frame, diff)

            # Display the resulting frame
            ostream((prev_frame, frame, mask, masked), (2, 2))
````

````python
frame_differencing_with_delay(cvu.cam_stream(), cvu.window_stream(), threshold=0.07, delay=15)
````

````python
frame_differencing_with_delay(cvu.dir_stream("./sequence-01", scale=0.7), cvu.window_stream(), threshold=0.11, delay=15)
````

# Three Frame Differencing

![](./three-frame-differencing.png)

$$
D(K) = || I(t) - I(t-K) || - || I(t) - I(t+K) ||
$$

````python
def three_frame_differencing(in_stream, out_stream, threshold=0.2, threeway_range=50):
    thresh = threshold * 3 * 255
    next_index = -1

    frames = [None] * (2*threeway_range + 1)
    indices = [None] * (2*threeway_range + 1)

    with in_stream as stream, out_stream as ostream:
        for ret, frame in stream:
            next_index += 1

            # Our operations on the frame come here
            next_i = next_index % (2*threeway_range + 1)
            normal_frame = frame
            frame = frame.astype(np.float32)
            frames[next_i] = frame
            indices[next_i] = next_index

            curr_index = max(0, next_index - threeway_range)
            curr_i = curr_index % (2*threeway_range + 1)
            curr = indices[curr_i]

            prev_index = max(0, next_index - 2*threeway_range)
            prev_i = prev_index % (2*threeway_range + 1)
            prev = indices[prev_i]

            diff_prev = np.linalg.norm((frames[prev_i] - frames[curr_i]), axis=2) > thresh
            diff_next = np.linalg.norm((frames[curr_i] - frames[next_i]), axis=2) > thresh
            diff = np.logical_and(diff_prev, diff_next).astype(np.float32)
            alldiff = np.logical_or(diff_prev, diff_next).astype(np.float32)
            mask_all = cvu.calc_mask(alldiff)
            mask_prev = cvu.calc_mask(diff_prev)
            mask_next = cvu.calc_mask(diff_next)
            
            mask, masked = cvu.create_mask(frames[curr_i], diff)

            ostream((frame, masked, mask, mask_prev, mask_next, mask_all), (2, 3))
````

````python
three_frame_differencing(cvu.cam_stream(), cvu.window_stream(), threshold=0.09, threeway_range=5)
````

````python
three_frame_differencing(cvu.dir_stream("./sequence-01", scale=0.7), cvu.window_stream(), threeway_range=15)
````

* delay too small: movement of slow object not detected (second 10)
* delay too big: extreme ghosting leads to good detections (second 1)
* cars & pedestrians move with different speeds (but just a single delay parameter ... )

````python
def three_frame_differencing_motion_history(in_stream, out_stream, threshold=0.2, threeway_range=50, history_keep=0.9):
    keep = 255 * history_keep
    thresh = threshold * 3 * 255
    next_index = -1

    frames = [None] * (2*threeway_range + 1)
    indices = [None] * (2*threeway_range + 1)
    
    H = None
    Z = None

    with in_stream as stream, out_stream as ostream:
        for ret, frame in stream:
            if H is None:
                H = np.zeros_like(frame).astype(np.float32)
                Z = np.zeros_like(frame).astype(np.float32)
            
            next_index += 1

            # Our operations on the frame come here
            next_i = next_index % (2*threeway_range + 1)
            normal_frame = frame
            frame = frame.astype(np.float32)
            frames[next_i] = frame
            indices[next_i] = next_index

            curr_index = max(0, next_index - threeway_range)
            curr_i = curr_index % (2*threeway_range + 1)
            curr = indices[curr_i]

            prev_index = max(0, next_index - 2*threeway_range)
            prev_i = prev_index % (2*threeway_range + 1)
            prev = indices[prev_i]

            diff_prev = np.linalg.norm((frames[prev_i] - frames[curr_i]), axis=2) > thresh
            diff_next = np.linalg.norm((frames[curr_i] - frames[next_i]), axis=2) > thresh
            diff = np.logical_and(diff_prev, diff_next).astype(np.float32)
            diff_all = np.logical_or(diff_prev, diff_next).astype(np.float32)
            mask_all = cvu.calc_mask(diff_all)
            mask_prev = cvu.calc_mask(diff_prev)
            mask_next = cvu.calc_mask(diff_next)
            
            mask, masked = cvu.create_mask(frames[curr_i], diff)
            
            tmp = np.maximum(H - (255 - keep), 1)
            H = np.maximum(mask, tmp)

            ostream((masked, mask, H, mask_prev, mask_next, mask_all), (2, 3))
````

````python
three_frame_differencing_motion_history(cvu.cam_stream(), cvu.window_stream(), threeway_range=15, history_keep=0.99)
````

````python
three_frame_differencing_motion_history(cvu.dir_stream("./sequence-01", scale=0.7), cvu.window_stream(), threeway_range=15, history_keep=0.99)
````

# Adaptive Background Subtraction

![](./adaptive-background-subtraction.png)

````python
def adaptive_background_subtraction(in_stream, out_stream, threshold=0.2, alpha = 0.2):
    thresh = threshold * 3 * 255
    prev_frame = None
    with in_stream as stream, out_stream as ostream:
        for ret, frame in stream:

            # Our operations on the frame come here
            frame = frame
            normal_frame = frame
            frame = frame.astype(np.float32)

            if prev_frame is None:
                prev_frame = frame

            diff = (np.linalg.norm((prev_frame - frame), axis=2) > thresh).astype(np.float32)
            prev_frame = alpha * frame + (1-alpha) * prev_frame
            
            mask, masked = cvu.create_mask(frame, diff)

            ostream([masked, mask], (1, 2))
````

````python
adaptive_background_subtraction(cvu.cam_stream(), cvu.window_stream(), threshold=0.1, alpha=0.2)
````

# Statistical Background Models
Model the background distribution of each pixel by a single Gaussian
centered at the mean pixel value:

Test if a newly observed pixel value has a high likelihood
under this Gaussian model. If yes then it is background.

````python
def gaussian_background(in_stream, out_stream, threshold=10, alpha = 0.2):
    mean = None
    sd = None

    with in_stream as stream, out_stream as ostream:
        _, first_im = next(stream)
        x, y, z = first_im.shape
        data_points = np.zeros((x, y, z, 101))
        data_points[:,:,:,0] = first_im.astype(np.float64)
        for i in range(100):
            ret, frame = next(stream)
            data_points[:,:,:,i+1] = frame.astype(np.float64)
            
        sd = np.var(data_points, axis=3)
        mean = np.mean(data_points, axis=3)
          
        print(np.min(sd), np.max(sd), np.average(sd))
        
        for ret, frame in stream:

            # Operations here
            original_frame = frame
            frame = frame.astype(np.float64)
 
            diff = np.abs(frame - mean) > (threshold * np.sqrt(sd))
            diff = np.max(diff, axis=2)

            mask, masked = cvu.create_mask(frame, diff)
            
#             print(np.min(sd), np.max(sd), np.average(sd))

            # Display the resulting frame
            ostream([masked, mask, mean, sd], (2, 2))
````

````python
gaussian_background(cvu.dir_stream("./sequence-01", scale=0.7), cvu.window_stream(), threshold=10, alpha=0.001)
````

````python
def gaussian_background_EMAF(in_stream, out_stream, threshold=10, alpha = 0.2):
    mean = None
    sd = None

    with in_stream as stream, out_stream as ostream:
        for ret, frame in stream:

            # Operations here
            original_frame = frame
            frame = frame.astype(np.float64)

            if mean is None:
                mean = frame
                sd = np.ones_like(frame).astype(np.float64)
 
            diff = np.abs(frame - mean) > (threshold * np.sqrt(sd))
            diff = np.max(diff, axis=2)
            mean = (1-alpha)*mean + alpha*frame
            sd = (1-alpha)*sd + alpha*(np.square(frame-mean))
            mask, masked = cvu.create_mask(frame, diff)

            # Display the resulting frame
            ostream([masked, mask, mean, sd], (2, 2))
````

````python
gaussian_background_EMAF(cvu.cam_stream(), cvu.window_stream(), threshold=5, alpha=0.01)
````

## Stauffer-Grimson (Adaptive Mixture Models)


````python
def gstauffer_grimson(in_stream, out_stream, threshold=1, alpha = 0.1, K=3, nr_standard_deviations = 2.5):
    mean = None
    sd = None
    prior = None
    high_variance = 756

    with in_stream as stream, out_stream as ostream:
        for ret, frame in stream:

            # Operations here
            original_frame = frame
            frame = frame.astype(np.float64)
            
            w, h, d = frame.shape
            
            # Initialization
            if mean is None:
                mean = np.random.uniform(0, 255, (K, w, h, d))
                sd = np.random.uniform(20, 20, (K, w, h))
                prior = np.ones((K, w, h)) / K
                
            prior_diff = prior / sd
            #prior_diff = np.sum(prior, axis=0)

#             print(np.min(prior_diff), np.max(prior_diff), np.average(prior_diff))
            background_mask = prior_diff > threshold
#             print(np.min(prior_diff), np.max(prior_diff), np.average(prior_diff))
                
            # Expand dimensions of lower dim objects.
            k_frame = np.repeat(frame[np.newaxis, :, :, :], K, axis=0)
            sd3 = np.repeat(sd[:, :, :, np.newaxis], d, axis=3)
                
            diff = np.abs(k_frame - mean)
            
            # Matched
            matched_channels = diff < (nr_standard_deviations * np.sqrt(sd3)) # all 4 dimensions
            matched_pixels = np.min(matched_channels, axis=3) # K, w, h
            pixel_component_mask = matched_pixels.cumsum(axis=0).cumsum(axis=0) == 1 # K, w, h
            pixel_component_mask_full = np.repeat(pixel_component_mask[:, :, :, np.newaxis], d, axis=3) # K, w, h , d
            pixel_mask = np.max(pixel_component_mask_full, axis=0) # w, h, d
            filter_mask = np.repeat(pixel_mask[np.newaxis, :, :, :], K, axis=0) # K, w, h, d
            
            background_maks = pixel_component_mask & background_mask
            background_mask = np.max(background_maks, axis=0)
            foreground_mask = np.invert(background_mask)
            #print(background_mask.shape, np.sum(background_mask), np.sum(foreground_mask), w*h)
            
            # Not matched
#             total_diff = np.linalg.norm(diff, axis=3) # K, w, h
#             max_diff = np.max(total_diff, axis=0, keepdims=True) == total_diff # K, w, h
#             diff_component_mask = max_diff.cumsum(axis=0).cumsum(axis=0) == 1 # K, w, h
#             diff_component_mask_full = np.repeat(diff_component_mask[:, :, :, np.newaxis], d, axis=3) # K, w, h , d
#             diff_component_mask_full = diff_component_mask_full & np.invert(filter_mask) # K, w, h, d
#             diff_pixel_mask = np.max(diff_component_mask_full, axis=3) # K, w, h
            min_prior = np.min(prior, axis=0, keepdims=True) == prior # K, w, h
            prior_component_mask = min_prior.cumsum(axis=0).cumsum(axis=0) == 1 # K, w, h
            prior_component_mask_full = np.repeat(prior_component_mask[:, :, :, np.newaxis], d, axis=3) # K, w, h , d
            prior_component_mask_full = prior_component_mask_full & np.invert(filter_mask) # K, w, h, d
            prior_pixel_mask = np.max(prior_component_mask_full, axis=3) # K, w, h

            # Update
            rho = stats.norm.pdf(k_frame, loc=mean, scale=sd3)
            rho = np.prod(rho, axis=3)
#             rho[rho > 1.0] = 1
            rho = alpha * rho
            rho_extended = np.repeat(rho[:, : , :, np.newaxis], 3, axis=3)
            new_mean = (1 - rho_extended) * mean + rho_extended * k_frame
            mean[pixel_component_mask_full] = new_mean[pixel_component_mask_full]
            
            dff = np.sum(np.abs(k_frame - mean), axis=3)
            new_sd = (1 - rho) * sd + rho * dff
            sd[pixel_component_mask] = new_sd[pixel_component_mask]
            #print(np.sum(diff_component_mask_full)/3, np.sum(pixel_component_mask_full), w * h)
            
            mean[prior_component_mask_full] = k_frame[prior_component_mask_full]
            sd[prior_pixel_mask] = high_variance
            
            prior = (1 - alpha) * prior + alpha * pixel_component_mask
            prior[prior_pixel_mask] = 0.0001
            prior_sum = np.sum(prior, axis=0)
            prior_sum = np.repeat(prior_sum[np.newaxis, :, :], K, axis=0)
            prior = prior / prior_sum
            
#             background_pixels = np.sum(np.square(k_frames - mean), axis=3) > (threshold * np.sqrt(sd))
#             background_pixels = np.average(background_pixels, axis=0)
            
            mask, masked = cvu.create_mask(frame, foreground_mask)
            min_per_pixel = np.min(sd, axis=0)
            sd_0 = cvu.calc_mask(sd[0])
            sd_1 = cvu.calc_mask(sd[1])
            sd_2 = cvu.calc_mask(sd[2])
            
            prior_0 = cvu.calc_mask(prior[0])
            prior_1 = cvu.calc_mask(prior[1])
            prior_2 = cvu.calc_mask(prior[2])
#             print(np.min(prior), np.max(prior), np.average(prior))
            #print(np.min(rho), np.max(rho), np.average(rho))
#             #print(np.min(dff), np.max(dff), np.average(dff))
#             print(np.min(sd), np.max(sd), np.average(sd))
            ostream([frame, mask, masked, mean[0], mean[1], mean[2], sd_0, sd_1, sd_2, prior_0, prior_1, prior_2], (4, 3))
````

**Properties**

Static foreground objects can be integrated into the mixture
* Advantage: This doesn’t destroy the existing background model.
* If an object is stationary for some time and then moves again,
the distribution for the background still exists
    * Quick recovery from such situations.
    
**Ordering of components by $w_k / \sigma_k$**
* Favors components that have more evidence (higher w k )
and a smaller variance (lower 3⁄4 k ).
* Those are typically the best candidates for background.

**Model can adapt to the complexity of the observed distribution.**
* If the distribution is unimodal, only a single component will be selected for
the background.
* This can be used to save memory and computation

## Kernel Background Modeling

````python
def kernel_background_modeling(in_stream, out_stream, threshold=10e-15, sigma = 11, N = 50):
    lut = stats.norm.pdf(np.arange(0, 256), loc=0, scale=sigma)
#     print(lut)

    with in_stream as stream, out_stream as ostream:
        ret, first_frame = next(stream)
        first_frame = first_frame.astype(np.int32)
        history = np.repeat(first_frame[np.newaxis, :, :, :], N, axis=0) # K, w, h , d
        k = 1
        i = 1
        for ret, frame in stream:
            frame = frame.astype(np.int32)
#             frames = np.repeat(frame[np.newaxis, :, :, :], k, axis=0) # K, w, h , d
            frames = np.expand_dims(frame, axis=0)
            if k < N:
                h = history[:k, :, :, :]
                k = k + 1
            else:
                h = history
                
#             print(frames.shape)
#             print(history.shape)
            probability_mask = np.take(lut, np.abs(h - frames)).prod(axis=3).sum(axis=0) / k
#             print(probability_mask.shape)
            print(np.min(probability_mask), np.max(probability_mask), np.average(probability_mask))
            foreground_mask = probability_mask < threshold
            
            history[i % N] = frame
            i = i + 1
            
            mask, masked = cvu.create_mask(frame, foreground_mask)
            
            ostream([frame, mask, masked], (1, 3))
````

````python
kernel_background_modeling(cvu.cam_stream(scale=0.5), cvu.window_stream(), N=10, sigma=20, threshold=1e-7)
````

````python
kernel_background_modeling(cvu.dir_stream("./sequence-01", scale=0.4), cvu.window_stream(), N=100, sigma=20, threshold=1e-7)
````
